### Модель

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor as RFR
from sklearn.model_selection import GridSearchCV
```

```python
params={
        'min_samples_leaf':[3,5,7,10],
        'n_estimators':[25,50,75],
        'max_depth':[3,5,7]    
        }
regr=GridSearchCV(RFR(),param_grid=params,cv=2)
```

```python
getto_feats=['Rooms', 'Square', 'HouseYear','Shops_1','Social_1','Helthcare_2','price_per_meter',
             'Shops_2_A','Shops_2_B','First_floor_0','First_floor_1']
```

```python
getto = data.loc[data.Social_2 <= 2500]
goods = data.loc[data.Social_2 > 2500]
```

```python
getto_train,getto_valid=train_test_split(getto, test_size=0.2, random_state=2)
goods_train,goods_valid=train_test_split(goods, test_size=0.2, random_state=2)
```

```python
# вычисляем среднею ценю за квадратный метр по району
```

```python
getto_mean = getto_train.groupby(['DistrictId'],as_index=False).agg({'Price':'mean', 'Square': 'mean'}).rename(columns={'Price': 'Mean_price', 'Square': 'Mean_square'})
goods_mean = goods_train.groupby(['DistrictId'],as_index=False).agg({'Price':'mean', 'Square': 'mean'}).rename(columns={'Price': 'Mean_price', 'Square': 'Mean_square'})
```

```python
getto_mean['price_per_meter']=getto_mean['Mean_price']/getto_mean['Mean_square']
goods_mean['price_per_meter']=goods_mean['Mean_price']/goods_mean['Mean_square']
```

```python
# мерджим
```

```python
getto_train=pd.merge(getto_train, getto_mean, on=['DistrictId'], how='left')
goods_train=pd.merge(goods_train, goods_mean, on=['DistrictId'], how='left')
```

```python
getto_valid=pd.merge(getto_valid, getto_mean, on=['DistrictId'], how='left')
goods_valid=pd.merge(goods_valid, goods_mean, on=['DistrictId'], how='left')
```

```python
# заполняем пропуски
```

```python
getto_train['price_per_meter']=getto_train['price_per_meter'].fillna(getto_train['price_per_meter'].mean())
goods_train['price_per_meter']=goods_train['price_per_meter'].fillna(goods_train['price_per_meter'].mean())
getto_valid['price_per_meter']=getto_valid['price_per_meter'].fillna(getto_valid['price_per_meter'].mean())
goods_valid['price_per_meter']=goods_valid['price_per_meter'].fillna(goods_valid['price_per_meter'].mean())
```

```python
# модель
```

```python
y_getto_train=getto_train['Price']
y_goods_train=goods_train['Price']
y_getto_valid=getto_valid['Price']
y_goods_valid=goods_valid['Price']
```

```python
getto_train=pd.get_dummies(getto_train)
goods_train=pd.get_dummies(goods_train)
getto_valid=pd.get_dummies(getto_valid)
goods_valid=pd.get_dummies(goods_valid)
```

```python
### лес на тех же данных
```

```python
%%time
regr.fit(getto_train.loc[:, getto_feats], y_getto_train)
```

```python
regr.best_params_
```

```python
model_getto_rfr=RFR(max_depth=7,min_samples_leaf=5,n_estimators=25)
```

```python
model_getto_rfr.fit(getto_train.loc[:, getto_feats], y_getto_train)
```

```python
getto_valid['Price'] = model_getto_rfr.predict(getto_valid.loc[:, getto_feats])
```

```python
r2_score(y_getto_valid, getto_valid['Price'])
```

```python
%%time
regr.fit(goods_train.loc[:, getto_feats], y_goods_train)
```

```python
regr.best_params_
```

```python
model_goods_rfr=RFR(max_depth=7,min_samples_leaf=7,n_estimators=75)
```

```python
model_goods_rfr.fit(goods_train.loc[:, getto_feats], y_goods_train)
```

```python
goods_valid['Price'] = model_goods_rfr.predict(goods_valid.loc[:, getto_feats])
```

```python
r2_score(y_goods_valid, goods_valid['Price'])
```

```python
final = getto_valid[['Id', 'Price']].append(goods_valid[['Id', 'Price']])
```
