```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('./creditcard.csv')
```

```python
class_counts = data['Class'].value_counts()
class_counts
```

```python
class_counts.plot(kind='bar')
plt.show()
```

```python
class_counts.plot(kind='bar', logy=True)
plt.show()
```

```python
v1_fraud = data['V1'].loc[(data['Class'] == 1)]
v1_nonfraud = data['V1'].loc[(data['Class'] == 0)]
```

```python
v1_fraud.hist(density = True, bins = 20, alpha = 0.5, color = 'Red', label = 'Class 1')
v1_nonfraud.hist(density = True, bins = 20,  alpha = 0.5, color = 'Grey', label = 'Class 2')
plt.legend(loc='upper left')
plt.xlabel('Class')
plt.show()
```
